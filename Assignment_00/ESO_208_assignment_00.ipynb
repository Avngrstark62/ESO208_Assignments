3. Write a function to print out the bill of three commodities which are given as 500 g of oranges
costing Rs.300, 250 g of grapes costing Rs.230 and 1 kg of apples costing Rs.215.


```python
def print_bill():
    # Define the commodities and their details
    commodities = [
        {"name": "Oranges", "weight": "500 g", "cost": 300},
        {"name": "Grapes", "weight": "250 g", "cost": 230},
        {"name": "Apples", "weight": "1 kg", "cost": 215}
    ]

    # Print the Bill
    print("Bill:")
    print("-" * 30)
    print(f"{'Commodity':<15} {'Weight':<10} {'Cost (Rs.)':>10}")
    print("-" * 30)

    for item in commodities:
        print(f"{item['name']:<15} {item['weight']:<10} {item['cost']:>10}")

    total_cost = sum(item["cost"] for item in commodities)
    print("-" * 30)
    print(f"{'Total Cost':<15} {'':<10} {total_cost:>10}")

# Call the function to print the bill
print_bill()
```

6. Create two random matrices, with values ranging from 1: 500, of size 1000×1000 and 100×100.
Use a for loop to compute the square of each element greater than 100 and the cube if the element
is less than 100.

```python
# Create two random matrices with values ranging from 1 to 500
matrix_1000x1000 = np.random.randint(1, 501, size=(1000, 1000))
matrix_100x100 = np.random.randint(1, 501, size=(100, 100))

# Function to process the matrix
def process_matrix(matrix):
    result_matrix = np.zeros_like(matrix)  # Create an empty matrix of the same size
    rows, cols = matrix.shape

    for i in range(rows):
        for j in range(cols):
            if matrix[i, j] > 100:
                result_matrix[i, j] = matrix[i, j] ** 2
            else:
                result_matrix[i, j] = matrix[i, j] ** 3

    return result_matrix

# Process both matrices
processed_matrix_1000x1000 = process_matrix(matrix_1000x1000)
processed_matrix_100x100 = process_matrix(matrix_100x100)

# Display a portion of the results for verification
print("Processed matrix (1000x1000):")
print(processed_matrix_1000x1000)

print("\nProcessed matrix (100x100):")
print(processed_matrix_100x100)
```

7. Write a function for factorial using while loop. Find the factorial of the prime number between 0
to 100 and store it in a list.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    result = 1
    while n > 1:
        result *= n
        n -= 1
    return result

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_factorials = []
for num in range(101):
    if is_prime(num):
        prime_factorials.append(factorial(num))

print("Factorials of prime numbers between 0 and 100:")
print(prime_factorials)
```

8. The Singular Value Decomposition of m × n matrix A is given by the formula:
A = UΣV
T
where:
• U: m × n matrix of the orthonormal eigenvectors of AAT
.
• V
T
: transpose of a n × n matrix containing the orthonormal eigenvectors of AT A.
• Σ: n × n diagonal matrix of the singular values which are the square roots of the eigenvalues
of AT A.
Define a 3 × 4 matrix of random numbers between 0 and 100. Perform SVD Analysis and print
U, V
T
, and Σ. Check the orthogonality of U and V . Compute the pseudo-inverse of A from the
SVD.


```python
# Defining a 3x4 matrix of random numbers between 0 and 100
A = np.random.rand(3, 4) * 100

# Performing SVD
U, Sigma, VT = np.linalg.svd(A)

# Creating the Σ matrix (Sigma is returned as a 1D array of singular values)
Sigma_matrix = np.zeros((3, 4))
np.fill_diagonal(Sigma_matrix, Sigma)

# Printing U, V^T, and Σ
print("Matrix U:")
print(U)

print("\nMatrix Σ:")
print(Sigma_matrix)

print("\nMatrix V^T:")
print(VT)

# Checking orthogonality of U and V
U_orthogonal = np.allclose(np.dot(U.T, U), np.eye(U.shape[1]))
VT_orthogonal = np.allclose(np.dot(VT, VT.T), np.eye(VT.shape[0]))

print("\nIs U orthogonal?", U_orthogonal)
print("Is V^T orthogonal?", VT_orthogonal)

# Computing the pseudo-inverse of A using the SVD components
Sigma_inv = np.zeros((4, 3))
np.fill_diagonal(Sigma_inv, 1 / Sigma)

A_pseudo_inverse = np.dot(np.dot(VT.T, Sigma_inv), U.T)

print("\nPseudo-inverse of A:")
print(A_pseudo_inverse)
```

8. Visualize the monthly average of temperature data for the last 100 years for the two different cities
- Kanpur and New Delhi (using their corresponding csv files). Compare the statistics of the two
cities. Use the following link for accessing the data: Kanpur and New Delhi.

```python
delhi_data = pd.read_csv('New_Delhi_Temp-TAVG-Trend_Copy.csv')
kanpur_data = pd.read_csv('Kanpur-TAVG-Trend_Copy.csv')
```

```python
# Preprocess the data: Create a Date column and set it as the index
def preprocess_data(df):
    df.dropna(subset=['Year', 'Month', 'Monthly anomalies'], inplace=True)
    df['Year'] = df['Year'].astype(int)
    df['Month'] = df['Month'].astype(int)
    df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str), format='%Y-%m')
    df.set_index('Date', inplace=True)
    return df
```

```python
kanpur_data = preprocess_data(kanpur_data)
delhi_data = preprocess_data(delhi_data)
```

```python
# Filter data for the last 100 years
start_date = pd.to_datetime('1924-01-01')
end_date = pd.to_datetime('2024-01-01')
kanpur_data_last_100_years = kanpur_data[(kanpur_data.index >= start_date) & (kanpur_data.index < end_date)]
delhi_data_last_100_years = delhi_data[(delhi_data.index >= start_date) & (delhi_data.index < end_date)]
```

```python
# Compute monthly averages
kanpur_monthly_avg = kanpur_data_last_100_years['Monthly anomalies'].groupby(kanpur_data_last_100_years.index.month).mean()
delhi_monthly_avg = delhi_data_last_100_years['Monthly anomalies'].groupby(delhi_data_last_100_years.index.month).mean()

# Plot the monthly averages
plt.figure(figsize=(12, 6))
plt.plot(kanpur_monthly_avg, label='Kanpur')
plt.plot(delhi_monthly_avg, label='New Delhi')
plt.title('Monthly Average Temperature Anomalies (Last 100 Years)')
plt.xlabel('Month')
plt.ylabel('Temperature Anomalies')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.show()

# Compare the statistics
kanpur_stats = kanpur_data_last_100_years['Monthly anomalies'].describe()
delhi_stats = delhi_data_last_100_years['Monthly anomalies'].describe()

print("Kanpur Temperature Anomalies Statistics (Last 100 Years):")
print(kanpur_stats)

print('')

print("New Delhi Temperature Anomalies Statistics (Last 100 Years):")
print(delhi_stats)
```
